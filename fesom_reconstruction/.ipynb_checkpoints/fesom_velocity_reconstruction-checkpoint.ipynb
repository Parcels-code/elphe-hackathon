# Cell-view linear reconstruction for triangular-mesh (FESOM) data

This notebook implements the **linear least-squares reconstruction** of section 2
(*"Cell view"*) of *"Simple reconstruction techniques for triangular meshes"*,
packages it as a Parcels interpolation kernel, and compares it against Parcels'
built-in unstructured-grid interpolators on a fictitious analytic field — the
same comparison style as the
[interpolation tutorial](../docs/user_guide/examples/tutorial_interpolation.ipynb).

In FESOM the horizontal velocities `U`, `V` are stored at the **centres of
triangles** (face-registered data). Parcels' default face interpolator
(`UxConstantFaceConstantZC`) is *piecewise constant*: every particle in a
triangle samples the same value and the field jumps across edges. The cell-view
reconstruction instead fits a **linear field inside each triangle**, with a slope
obtained by least squares from the neighbouring triangles.

The slope depends only on the *mesh geometry* — the relative positions of the
neighbouring cell centroids. These are the "additional metric terms" referred to
in the paper (`gradient_vec` in the FESOM source). We compute them once and
**store them in the `uxgrid` object** the fields hang off, then read them back
inside the interpolation kernel.


## The method (section 2)

Let $c$ index the triangles. The velocity $u_c$ is attached to the centroid
$\mathbf{x}_c$. The linear reconstruction on cell $c$ is

$$ U_c(\mathbf{x}) = u_c + (a_x, a_y)\cdot(\mathbf{x}-\mathbf{x}_c), \tag{1}$$

with the gradient $(a_x, a_y)$ chosen to minimise the misfit at the edge
neighbours $N(c)$,

$$ \mathcal{F} = \sum_{n\in N(c)}\big(u_n - u_c - (a_x,a_y)\cdot(\mathbf{x}_n-\mathbf{x}_c)\big)^2 .$$

Setting $\partial\mathcal{F}/\partial a_x = \partial\mathcal{F}/\partial a_y = 0$
gives

$$ a_x = \sum_{n\in N(c)} g^x_{cn}(u_n-u_c), \qquad
   a_y = \sum_{n\in N(c)} g^y_{cn}(u_n-u_c), $$

$$ g^x_{cn} = \frac{(x_n-x_c)\,Y^2 - (y_n-y_c)\,XY}{d}, \qquad
   g^y_{cn} = \frac{(y_n-y_c)\,X^2 - (x_n-x_c)\,XY}{d}, $$

$$ d = X^2 Y^2 - (XY)^2,\quad
   X^2 = \!\!\sum_{n\in N(c)}\!\!(x_n-x_c)^2,\;
   Y^2 = \!\!\sum_{n\in N(c)}\!\!(y_n-y_c)^2,\;
   XY = \!\!\sum_{n\in N(c)}\!\!(x_n-x_c)(y_n-y_c). $$

The coefficients $g^x_{cn}, g^y_{cn}$ are the metric terms; they are independent
of the data, so they are precomputed once and cached on the grid.

**Spherical metric.** The paper notes (concluding remarks) that in spherical
geometry the differences $\mathbf{x}_n-\mathbf{x}_c$ must be measured with the
local metric. We follow FESOM's "cosine on triangles" recipe: differences are
local **east/north distances in metres** relative to each cell centroid,
$\Delta x = (\lambda_n-\lambda_c)\cos\phi_c R_{\deg}$,
$\Delta y = (\phi_n-\phi_c) R_{\deg}$, with longitude differences cyclically
reduced and $R_{\deg}=1852\cdot 60$ m/deg (matching the unit conversion in
Parcels' `Ux_Velocity`). On a `"flat"` mesh the raw coordinate differences are
used. Because the reconstruction returns a value in the same units as the raw
face data, it drops straight into the existing kernels.


```python
from pathlib import Path

import matplotlib.pyplot as plt
import matplotlib.tri as mtri
import numpy as np
import xarray as xr
from uxarray.constants import INT_FILL_VALUE

import parcels
import parcels.tutorial
```

## The reconstruction kernel

Two ingredients, both mesh-aware (`"flat"` vs `"spherical"`):

1. `local_offsets` — the local planar offset of a point from a cell centroid.
2. `build_reconstruction_terms` — turns the centroid geometry into the
   coefficient arrays $g^x_{cn}, g^y_{cn}$ and **stores them on `uxgrid._ds`**
   together with the neighbour indices and a validity mask, so they travel with
   the grid object every field references.

`uxarray` gives us the edge-neighbour set $N(c)$ directly as
`uxgrid.face_face_connectivity`, shape `(n_face, n_max_face_faces)`, with absent
neighbours (boundary triangles) flagged by `INT_FILL_VALUE`.


```python
DEG2M = 1852.0 * 60.0  # metres per degree (matches Ux_Velocity unit conversion)


def local_offsets(mesh, lon_c, lat_c, lon, lat):
    # local planar offset (dx, dy) of (lon, lat) from centroid (lon_c, lat_c)
    if mesh == "spherical":
        dlon = ((np.asarray(lon) - lon_c + 180.0) % 360.0) - 180.0
        dx = dlon * np.cos(np.deg2rad(lat_c)) * DEG2M
        dy = (np.asarray(lat) - lat_c) * DEG2M
    else:  # flat mesh: coordinates are already a planar (x, y)
        dx = np.asarray(lon) - lon_c
        dy = np.asarray(lat) - lat_c
    return dx, dy


# Compute and attach cell-view least-squares gradient coefficients. Adds to
# uxgrid._ds: gradient_coeff_x/_y, recon_face_neighbors (fill -> self),
# recon_neighbor_mask (1.0 valid / 0.0 absent), all (n_face, n_max_face_faces).
def build_reconstruction_terms(uxgrid, mesh):
    nbr = np.asarray(uxgrid.face_face_connectivity.values)        # (F, K)
    flon = np.asarray(uxgrid.face_lon.values)
    flat = np.asarray(uxgrid.face_lat.values)

    valid = nbr != INT_FILL_VALUE
    nbr_safe = np.where(valid, nbr, 0)

    dx, dy = local_offsets(mesh, flon[:, None], flat[:, None], flon[nbr_safe], flat[nbr_safe])
    dx = np.where(valid, dx, 0.0)                                 # absent -> no contribution
    dy = np.where(valid, dy, 0.0)

    X2 = np.sum(dx * dx, axis=1)
    Y2 = np.sum(dy * dy, axis=1)
    XY = np.sum(dx * dy, axis=1)
    d = X2 * Y2 - XY * XY

    # cells with < 2 independent neighbours give d == 0 -> flat (zero-gradient)
    # fallback, i.e. the original piecewise-constant value.
    ok = d != 0.0
    inv_d = np.where(ok, 1.0 / np.where(ok, d, 1.0), 0.0)

    gx = (dx * Y2[:, None] - dy * XY[:, None]) * inv_d[:, None]
    gy = (dy * X2[:, None] - dx * XY[:, None]) * inv_d[:, None]
    gx = np.where(valid, gx, 0.0)
    gy = np.where(valid, gy, 0.0)

    dims = ("n_face", "n_max_face_faces")
    uxgrid._ds["gradient_coeff_x"] = xr.DataArray(gx, dims=dims)
    uxgrid._ds["gradient_coeff_y"] = xr.DataArray(gy, dims=dims)
    uxgrid._ds["recon_face_neighbors"] = xr.DataArray(nbr_safe, dims=dims)
    uxgrid._ds["recon_neighbor_mask"] = xr.DataArray(valid.astype(np.float64), dims=dims)
    return uxgrid
```

A Parcels scalar interpolator has the signature
`(particle_positions, grid_positions, field)`. Ours reads $u_c$ and the
neighbour values $u_n$ from `field.data`, combines them with the cached
coefficients to form the gradient $(a_x, a_y)$, and evaluates equation (1) at the
particle's offset from the centroid. Vertically it stays piecewise-constant per
layer, matching the `...ConstantZC` interpolator it replaces.


```python
# Cell-view linear least-squares reconstruction for face-registered, layer-centered data.
def LinearFaceRecon(particle_positions, grid_positions, field):
    ti = grid_positions["T"]["index"]
    zi = grid_positions["Z"]["index"]
    fi = grid_positions["FACE"]["index"]

    uxg = field.grid.uxgrid
    gx = uxg._ds["gradient_coeff_x"].values[fi]          # (M, K)
    gy = uxg._ds["gradient_coeff_y"].values[fi]
    nb = uxg._ds["recon_face_neighbors"].values[fi]       # (M, K)
    mask = uxg._ds["recon_neighbor_mask"].values[fi]

    uc = field.data.values[ti, zi, fi]                                  # (M,)
    un = field.data.values[ti[:, None], zi[:, None], nb]               # (M, K)
    du = (un - uc[:, None]) * mask
    ax = np.sum(gx * du, axis=1)
    ay = np.sum(gy * du, axis=1)

    flon = uxg.face_lon.values[fi]
    flat = uxg.face_lat.values[fi]
    dx, dy = local_offsets(field.grid._mesh, flon, flat, particle_positions["lon"], particle_positions["lat"])
    return uc + ax * dx + ay * dy
```

## Comparison on a fictitious analytic field

Following the [interpolation tutorial](../docs/user_guide/examples/tutorial_interpolation.ipynb),
we use the small flat Delaunay dataset that ships with Parcels. It carries a
fictitious tracer

$$ T(x, y) = \sin(\pi y)\cos(\pi x) $$

sampled both at face centres (`T_face`) and at nodes (`T_node`). Because we know
$T$ analytically, we can measure the error of each interpolation method exactly.


```python
from parcels._datasets.unstructured.generated import simple_small_delaunay

ds = simple_small_delaunay(nx=24, ny=24)
fieldset = parcels.FieldSet.from_ugrid_conventions(ds, mesh="flat")

# attach the metric terms to the grid shared by the fields
uxgrid = fieldset.fields["T_face"].grid.uxgrid
build_reconstruction_terms(uxgrid, mesh="flat")
print("default T_face interpolator:", fieldset.fields["T_face"].interp_method.__name__)
print("stored on uxgrid._ds:", [v for v in uxgrid._ds.data_vars if v.startswith(("gradient", "recon"))])
```

### Quick geometric sanity check

The metric terms are purely geometric, so feeding *random* neighbour values
through the stored coefficients must match an explicit `numpy.linalg.lstsq` solve
of the same over-determined system $\Delta\mathbf{x}_n\cdot(a_x,a_y)=u_n-u_c$.


```python
rng = np.random.default_rng(0)
flon, flat = uxgrid.face_lon.values, uxgrid.face_lat.values
nbr_safe = uxgrid._ds["recon_face_neighbors"].values
mask = uxgrid._ds["recon_neighbor_mask"].values
gx = uxgrid._ds["gradient_coeff_x"].values
gy = uxgrid._ds["gradient_coeff_y"].values

du_rand = rng.standard_normal(nbr_safe.shape) * mask
ax_coeff = np.sum(gx * du_rand, axis=1)
ay_coeff = np.sum(gy * du_rand, axis=1)

max_err = 0.0
for c in np.where(mask.sum(axis=1) >= 2)[0]:
    use = mask[c] > 0
    nb = nbr_safe[c][use]
    A = np.column_stack([flon[nb] - flon[c], flat[nb] - flat[c]])  # flat mesh -> raw offsets
    sol, *_ = np.linalg.lstsq(A, du_rand[c][use], rcond=None)
    max_err = max(max_err, abs(sol[0] - ax_coeff[c]), abs(sol[1] - ay_coeff[c]))
print(f"max |coeff - lstsq| gradient error = {max_err:.3e}")
assert max_err < 1e-6
```

### Sampling the tracer with each interpolator

Exactly as in the interpolation tutorial: a `SampleParticle` carries a `tracer`
variable, a one-step kernel writes the interpolated value onto it, and we swap
`fieldset.T_face.interp_method` between methods. We add our
`LinearFaceRecon` to the two built-ins:

- `UxConstantFaceConstantZC` — piecewise constant (the default for face data),
- `LinearFaceRecon` — our cell-view linear reconstruction,
- `UxLinearNodeConstantZC` — barycentric-linear, but it needs *node* data
  (`T_node`); shown as the accuracy reference a linear scheme can reach.


```python
from parcels.interpolators import UxConstantFaceConstantZC, UxLinearNodeConstantZC

SampleParticle = parcels.Particle.add_variable(parcels.Variable("tracer", dtype=np.float64, initial=np.nan))


def SampleFace(particles, fieldset):
    particles.tracer = fieldset.T_face[particles]


def SampleNode(particles, fieldset):
    particles.tracer = fieldset.T_node[particles]


# sample points well inside the [0, 1] x [0, 1] domain (avoid boundary triangles)
xv, yv = np.meshgrid(np.linspace(0.15, 0.85, 22), np.linspace(0.15, 0.85, 22))
px, py = xv.ravel(), yv.ravel()
T_exact = np.sin(np.pi * py) * np.cos(np.pi * px)


def sample(field_name, interp, kernel):
    fieldset.fields[field_name].interp_method = interp
    pset = parcels.ParticleSet(fieldset, pclass=SampleParticle, lon=px, lat=py)
    pset.execute(kernel, runtime=np.timedelta64(1, "D"), dt=np.timedelta64(1, "D"), verbose_progress=False)
    return np.asarray(pset.tracer)


results = {
    "Constant (face)": sample("T_face", UxConstantFaceConstantZC, SampleFace),
    "Linear recon (face)": sample("T_face", LinearFaceRecon, SampleFace),
    "Barycentric (node)": sample("T_node", UxLinearNodeConstantZC, SampleNode),
}

print(f"{'method':<22s} {'RMS error':>12s} {'max error':>12s}")
for name, vals in results.items():
    err = vals - T_exact
    print(f"{name:<22s} {np.sqrt(np.nanmean(err**2)):>12.3e} {np.nanmax(np.abs(err)):>12.3e}")
```

The linear reconstruction cuts the face-data error by a large factor over the
piecewise-constant default, approaching the accuracy of barycentric
interpolation — but using only the face-registered velocities FESOM actually
provides, with no node data required.

Let us visualise it: the tracer field in the background, the sampled particle
values as filled circles (top row), and the signed error (bottom row).


```python
x = uxgrid.node_lon.values
y = uxgrid.node_lat.values
triang = mtri.Triangulation(x, y, triangles=uxgrid.face_node_connectivity.values)
T_face_vals = np.squeeze(ds["T_face"][0, 0, :].values)

fig, axes = plt.subplots(2, 3, figsize=(15, 9), constrained_layout=True)
for j, (name, vals) in enumerate(results.items()):
    ax = axes[0, j]
    ax.tripcolor(triang, facecolors=T_face_vals, shading="flat", cmap="viridis", vmin=-1, vmax=1, alpha=0.6)
    ax.triplot(triang, color="0.6", linewidth=0.3)
    ax.scatter(px, py, c=vals, cmap="viridis", vmin=-1, vmax=1, edgecolors="k", s=30)
    ax.set_title(name)
    ax.set_aspect("equal", "box")

    axe = axes[1, j]
    sc = axe.scatter(px, py, c=vals - T_exact, cmap="bwr", vmin=-0.15, vmax=0.15, edgecolors="k", s=30)
    axe.set_title(f"{name}: error")
    axe.set_aspect("equal", "box")
fig.colorbar(sc, ax=axes[1, :], shrink=0.7, label="sampled - exact")
plt.show()
```

### Exactness on a linear field

A linear reconstruction must reproduce a genuinely linear field *exactly*. We
overwrite `T_face` with $T = 0.3 + 0.7x - 0.4y$ and re-sample: the linear
reconstruction recovers it to round-off, while the piecewise-constant scheme
retains an $O(\text{cell size})$ error.


```python
T_face_field = fieldset.fields["T_face"]
orig = T_face_field.data.values.copy()
T_face_field.data.values[...] = (0.3 + 0.7 * flon - 0.4 * flat)[None, None, :]
T_lin_exact = 0.3 + 0.7 * px - 0.4 * py

for name, interp in [("Constant (face)", UxConstantFaceConstantZC), ("Linear recon (face)", LinearFaceRecon)]:
    vals = sample("T_face", interp, SampleFace)
    err = np.nanmax(np.abs(vals - T_lin_exact))
    print(f"{name:<22s} max error on linear field = {err:.3e}")

T_face_field.data.values[...] = orig  # restore the analytic tracer
```

## Applying it to real FESOM velocities

Now the same kernel on a real FESOM2 snapshot — the periodic-channel dataset from
the [FESOM tutorial](../docs/user_guide/examples/tutorial_fesom.ipynb). We load
the data, build the metric terms on its (spherical) grid, attach the
reconstruction kernel to `U` and `V`, and advect particles with it. The
`Ux_Velocity` vector kernel calls our reconstruction for the horizontal
components and handles the m/s → deg/s conversion; `W` (node-registered) keeps
its barycentric-linear interpolator.


```python
import uxarray as ux

for name in [
    "FESOM_periodic_channel/fesom_channel",
    "FESOM_periodic_channel/u.fesom_channel",
    "FESOM_periodic_channel/v.fesom_channel",
    "FESOM_periodic_channel/w.fesom_channel",
]:
    parcels.tutorial.open_dataset(name)

from parcels._datasets.remote import _DATA_HOME

data_dir = Path(_DATA_HOME) / "data" / "FESOM_periodic_channel"
grid_path = str(data_dir / "fesom_channel.nc")
data_paths = [str(data_dir / f"{v}.fesom_channel.nc") for v in ["u", "v", "w"]]

fds = ux.open_mfdataset(grid_path, data_paths).rename_vars({"u": "U", "v": "V", "w": "W"})
fds = parcels.convert.fesom_to_ugrid(fds)
fesom = parcels.FieldSet.from_ugrid_conventions(fds, mesh="spherical")

build_reconstruction_terms(fesom.fields["U"].grid.uxgrid, mesh="spherical")
n_nbr = fesom.fields["U"].grid.uxgrid._ds["recon_neighbor_mask"].values.sum(axis=1)
print(f"FESOM grid: {fesom.fields['U'].grid.uxgrid.n_face} faces, "
      f"{(n_nbr == 3).sum()} interior (3-neighbour) triangles")
```

```python
def run_fesom(interp, label):
    fesom.fields["U"].interp_method = interp
    fesom.fields["V"].interp_method = interp
    lon_grid, lat_grid = np.meshgrid(np.linspace(0.5, 4.0, 10), np.linspace(3.0, 15.0, 4))
    pset = parcels.ParticleSet(
        fesom, pclass=parcels.Particle,
        lon=lon_grid.ravel(), lat=lat_grid.ravel(), z=np.full(lon_grid.size, 50.0),
    )
    out = parcels.ParticleFile(f"recon-{label}.parquet", outputdt=np.timedelta64(1, "h"), overwrite=True)
    pset.execute(
        [parcels.kernels.AdvectionRK4],
        runtime=np.timedelta64(2, "D"), dt=np.timedelta64(5, "m"),
        output_file=out, verbose_progress=False,
    )
    return parcels.read_particlefile(f"recon-{label}.parquet")


df_const = run_fesom(UxConstantFaceConstantZC, "const")
df_lin = run_fesom(LinearFaceRecon, "linear")
print("advection done")
```

```python
ug = fesom.fields["U"].grid.uxgrid
triang = mtri.Triangulation(ug.node_lon.values, ug.node_lat.values, triangles=ug.face_node_connectivity.values)
depth_idx = int(np.argmin(np.abs(fds.zc.values - 50.0)))
speed = np.hypot(
    np.asarray(fds["U"].isel(zc=depth_idx)).squeeze(), np.asarray(fds["V"].isel(zc=depth_idx)).squeeze()
)

fig, ax = plt.subplots(figsize=(11, 5))
ax.tripcolor(triang, facecolors=speed, shading="flat", cmap="Blues", alpha=0.7)
for df, color, lab in [(df_const, "tab:red", "piecewise constant"), (df_lin, "k", "linear reconstruction")]:
    first = True
    for traj in df.sort("time").partition_by("particle_id"):
        ax.plot(traj["lon"], traj["lat"], color=color, lw=0.8, alpha=0.8, label=lab if first else None)
        first = False
ax.set_xlabel("Longitude [deg E]")
ax.set_ylabel("Latitude [deg N]")
ax.set_title("FESOM trajectories: piecewise-constant vs cell-view linear reconstruction")
ax.legend(loc="upper right")
plt.show()
```

## Summary

- The cell-view reconstruction (section 2) turns the discontinuous,
  piecewise-constant FESOM face velocities into a **linear field per triangle**,
  fitted by least squares to the edge neighbours.
- The fit reduces to multiplying neighbour differences by purely **geometric
  coefficients** $g^x_{cn}, g^y_{cn}$ (the "metric terms"), computed once and
  **stored on the `uxgrid`** (`uxgrid._ds`) so they travel with the grid the
  fields reference.
- A custom kernel reads those cached terms and evaluates the linear
  reconstruction; it slots into the existing `Ux_Velocity` vector kernel
  unchanged.
- On a fictitious analytic field it sharply reduces the face-data interpolation
  error versus the piecewise-constant default and is **exact for linear fields**,
  while needing only the face-registered data FESOM provides.

Natural extensions: the *vertex view* (section 3) for a continuous linear field,
and the $P_2$ quadratic reconstruction (section 3.1) — both store larger per-cell
coefficient sets on the same `uxgrid` object.
